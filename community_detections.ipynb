```python

import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from igraph import *
```

```python
g_fb = read('./Bignetwork/source_graph/facebook.gml')
g_ins = read('./Bignetwork/source_graph/instagram.gml')
g_tw = read('./Bignetwork/source_graph/twitter.gml')
```

```python
g_fb.to_undirected()
c_fb = g_fb.community_label_propagation()

for i in range(len(c_fb.membership)):
    g_fb.vs[i]['community'] = c_fb.membership[i]

g_fb.write_gml('c_fb.gml')
```

```python
print(c_fb)
```

```python
g_fb.modularity(c_fb)
```
